```python
import matplotlib.pyplot as plt
import numpy as np
ap=np.array([1,5])
bp=np.array([4,10])
cp=np.array([9,120])
plt.plot(ap,bp,cp)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.array(['cse','math1','math2','eng','vedc'])
b=np.array([90,78,89,88,20])
plt.bar(a,b)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([10,15,20,14,18])
a=["maths","cse","eng","math1","math2"]
plt.pie(y,labels=a)
plt.show()
```
