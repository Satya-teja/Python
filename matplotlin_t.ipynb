```python
import matplotlib.pyplot as plt
import numpy as np

import random

a=[]

b=[]

for i in range(7):
    a.append(random.randint(1,10))

for i in range(7):
    b.append(random.randint(1,10))

a=np.array(a)

b=np.array(b)

plt.plot(a,b,linestyle='dotted')

plt.xlabel("Hours")

plt.ylabel("Subjects")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([3,8,1,10])

plt.plot(x,marker='o',linestyle='dotted')

plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as np

a=np.array([35,25,25,15])

label=['CSE',"MECH","CIVIL",'ECE']

plt.pie(a,labels=label)

plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as np

a=np.array([35,25,25,15])

e= [0.2, 0, 0, 0]
l=['Python','Chemistry','English','Maths']
plt.pie(a, labels = l, explode = e)
plt.show() 
```
